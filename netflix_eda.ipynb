```python
sns.countplot(x="type", data=df)
plt.title("Distribution of Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("TOP 10 Countries Producing Netflix Content")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries Producing Netflix Content")
plt.savefig("top_countries.png")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title("Netflix Content Release Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```


## Key Insights

1. Netflix hosts more Movies than TV Shows in its catalog.
2. The United States contributes the largest amount of content.
3. The number of titles released on Netflix increased significantly after 2015.
4. Streaming demand has driven rapid content expansion in recent years.
